```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.array([
    [0, 0],
    [3, 0],
    [3, 3],
    [0, 3],
]).T
B = np.array([
    [1, 3],
    [2, 1],
])
C = np.matmul(B, A)
A = np.column_stack((A, A[:, 0]))
C = np.column_stack((C, C[:, 0]))
fig, ax = plt.subplots()
plt.xlim(0, 13)
plt.ylim(0, 10)
plt.plot(A[0, :], A[1, :])
plt.plot(C[0, :], C[1, :])
plt.show()
```
