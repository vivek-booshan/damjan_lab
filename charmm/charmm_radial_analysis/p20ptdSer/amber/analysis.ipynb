```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
gofr_water = np.loadtxt('gofr_water.dat')
gofr_ions = np.loadtxt('gofr_ions.dat')
```

```python
plt.plot(gofr_water[:, 0], gofr_water[:, 1])
plt.title("water")
```

```python
plt.plot(gofr_water[:, 0], gofr_ions[:, 1]/gofr_water[:, 1])
plt.xlabel('distance')
plt.ylabel('RDF')
plt.title("normalized")
```

```python
plt.plot(gofr_ions[:, 0], gofr_ions[:, 1])
plt.title("ions")
```
