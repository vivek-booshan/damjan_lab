```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ptdSer = "ptdSer/amber/noxef/"
EF_ptdSer = "ptdSer/amber/xef/"

p20_gofr_water = np.loadtxt('p20' + ptdSer + "gofr_water.dat")
p20_gofr_ions = np.loadtxt('p20' + ptdSer + 'gofr_ions.dat')
p20_gofr_water_xef = np.loadtxt('p20' + EF_ptdSer + 'gofr_water.dat')
p20_gofr_ions_xef = np.loadtxt('p20' + EF_ptdSer + 'gofr_ions.dat')

p10_gofr_water = np.loadtxt('p10' + ptdSer + 'gofr_water.dat')
p10_gofr_ions = np.loadtxt('p10' + ptdSer + 'gofr_ions.dat')
p10_gofr_water_xef = np.loadtxt('p10' + EF_ptdSer + 'gofr_water.dat')
p10_gofr_ions_xef = np.loadtxt('p10' + EF_ptdSer + 'gofr_ions.dat')

p30_gofr_water = np.loadtxt('p30' + ptdSer + 'gofr_water.dat')
p30_gofr_ions = np.loadtxt('p30' + ptdSer + 'gofr_ions.dat')
p30_gofr_water_xef = np.loadtxt('p30' + ptdSer + 'gofr_water.dat')
p30_gofr_ions_xef = np.loadtxt('p30' + EF_ptdSer + 'gofr_ions.dat')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 8), sharey=True)

axes[0].plot(p20_gofr_water[:, 0], p20_gofr_ions[:, 1] / p20_gofr_water[:, 1], label='20% ptdSer')
axes[0].plot(p10_gofr_water[:, 0], p10_gofr_ions[:, 1] / p10_gofr_water[:, 1], label='10% ptdSer')
axes[0].plot(p30_gofr_water[:, 0], p30_gofr_ions[:, 1] / p30_gofr_water[:, 1], label='30% ptdSer')
axes[0].set(title='% ptdSer Without EF')
axes[0].legend(['20% ptdSer', '10% ptdSer', '30% ptdSer'])

axes[1].plot(p20_gofr_water_xef[:, 0], p20_gofr_ions_xef[:, 1] / p20_gofr_water_xef[:, 1])
axes[1].plot(p10_gofr_water_xef[:, 0], p10_gofr_ions_xef[:, 1] / p10_gofr_water_xef[:, 1])
axes[1].plot(p30_gofr_water_xef[:, 0], p30_gofr_ions_xef[:, 1] / p30_gofr_water_xef[:, 1])
axes[1].set(title='% ptdSer with EF')
axes[1].legend(['20% ptdSer', '10% ptdSer', '30% ptdSer'])

# for i, ax in enumerate(axes.flat):
#     ax.set(
# #         title=f"{(i+1)*10}% PtdSer with No E.F.",
# #         xlabel='distance (angstroms)',
# #     )
# axes[0].set(ylabel='RDF')
```

```python
plt.plot(p10_gofr_water[:, 0], p10_gofr_ions[:, 1]/p10_gofr_water[:, 1])
plt.xlabel('distance (angstroms)')
plt.ylabel('RDF')
plt.title("10% PtdSer in Electric Field")
```

```python
plt.plot(p30_gofr_water[:, 0], p30_gofr_ions[:, 1]/p30_gofr_water[:, 1])
plt.xlabel('distance (angstroms)')
plt.ylabel('RDF')
plt.title("30% PtdSer in Electric Field")
```
